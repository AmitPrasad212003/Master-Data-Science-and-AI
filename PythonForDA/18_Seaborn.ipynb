## 1Ô∏è‚É£ What is Seaborn?

**Seaborn** is a **high-level statistical data visualization library** built **on top of Matplotlib**.

üìå Seaborn works **best with Pandas DataFrames**

---

## 2Ô∏è‚É£ Why Use Seaborn? (REAL REASON)

Matplotlib problems:

- Too much code
- No default styling
- Manual legends, colors, grids

### ‚úÖ Seaborn solves this:

‚úî Beautiful default themes

‚úî Less code

‚úî Statistical plots

‚úî Automatic legends

‚úî Works directly with DataFrames

---

## 3Ô∏è‚É£ Importing Seaborn (MANDATORY)

```python

import matplotlib.pyplotas plt

```

üìå `plt.show()` is still required (because Seaborn uses Matplotlib internally)

## 4Ô∏è‚É£ Default Style in Seaborn (VERY IMPORTANT)

By default, Seaborn:

- Uses **whitegrid theme**
- Soft colors
- Automatic legend
- Grid ON

```python
sns.get_theme()

```

## 5Ô∏è‚É£ Basic Seaborn Plot (`lineplot`)

### Syntax

```python
sns.lineplot(x=x, y=y)
plt.show()

```

---

### Example

```python
x = [1,2,3,4]
y = [10,20,30,40]

sns.lineplot(x=x, y=y)
plt.show()

```

### Default Output

‚úî Smooth line

‚úî Blue color

‚úî Grid ON

‚úî Axis labels auto-detected

---

## 6Ô∏è‚É£ Seaborn with Pandas DataFrame (BEST USE)

```python
import pandasas pd

df = pd.DataFrame({
"Day": [1,2,3,4],
"Sales": [100,200,150,300]
})

sns.lineplot(data=df, x="Day", y="Sales")
plt.show()

```

üìå No need to manually extract columns

---

## 7Ô∏è‚É£ Multiple Lines in ONE LINE (IMPORTANT üî•)

### Syntax

```python

sns.lineplot(x=x, y=y2)

```

or (BEST WAY)

```python

```

---

### Example

```python
df = pd.DataFrame({
"x": [1,2,3,4,1,2,3,4],
"y": [10,20,30,40,15,25,35,45],
"Type": ["A","A","A","A","B","B","B","B"]
})

```

### Default Output

‚úî Multiple colored lines

‚úî Automatic legend

‚úî Smooth lines

````python
# Seaborn for plotting and styling
import seaborn as sns
df = sns.load_dataset('iris')
print(df.head())
````

````python
df.species.value_counts()
````

````python
sns.kdeplot(df['sepal_length'], shade=True)
````

````python
# KDE: kernel density estimation
for col in ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']:
    sns.kdeplot(df[col], shade=True)
````

#### Histograms + KDE = Distplot

````python
sns.distplot(df['sepal_length'])
````

````python
sns.pairplot(df, hue='species', size=2.5);
````
