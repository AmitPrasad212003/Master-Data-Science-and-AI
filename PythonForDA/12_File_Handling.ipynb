## 1Ô∏è‚É£ What is File Handling?

**File handling** allows a program to:

- **Create**
- **Read**
- **Write**
- **Update**
- **Delete**

data stored permanently on disk.

üìå Data in variables is **temporary (RAM)**

üìå Data in files is **persistent (storage)**

---

## 2Ô∏è‚É£ Why File Handling is Needed?

‚úî Store user data

‚úî Save logs

‚úî Read configuration files

‚úî Handle large datasets

‚úî Data persistence between program runs

## 3Ô∏è‚É£ Basic File Operations

| Operation | Method |
| --- | --- |
| Open file | `open()` |
| Read file | `read()`, `readline()`, `readlines()` |
| Write file | `write()`, `writelines()` |
| Close file | `close()` |

## 4Ô∏è‚É£ Opening a File

### Syntax

```python

```

### Example

```python

```

---

## 5Ô∏è‚É£ File Modes (VERY IMPORTANT)

| Mode | Meaning | File Exists | File Not Exists |
| --- | --- | --- | --- |
| `r` | Read | ‚úî | ‚ùå Error |
| `w` | Write | Overwrite | Create |
| `a` | Append | Write end | Create |
| `x` | Exclusive create | ‚ùå Error | Create |
| `r+` | Read + Write | ‚úî | ‚ùå |
| `w+` | Write + Read | Overwrite | Create |
| `a+` | Append + Read | ‚úî | Create |

## File handling Operations :
1Ô∏è‚É£ Creating & Writing to a File (`w` mode)

### Code

```python
f =open("sample.txt","w")
f.write("Hello Python\n")
f.write("File Handling")
f.close()

```

### File Content (`sample.txt`)

```
Hello Python
File Handling

```

üìå `w` mode **creates** file if not exists

üìå **overwrites** existing content

---

## 2Ô∏è‚É£ Reading Entire File (`read()`)

```

```

## 3Ô∏è‚É£ Reading Line by Line (`readline()`)

## 4Ô∏è‚É£ Reading All Lines as List (`readlines()`)

## 5Ô∏è‚É£ Looping Through File (BEST PRACTICE)

### Code

```python
withopen("sample.txt")as f:
for linein f:
print(line.strip())

```

### Output

```
Hello Python
File Handling

```

‚úî Memory efficient

‚úî Auto closes file

## 1Ô∏è‚É£5Ô∏è‚É£ JSON File Handling (REAL WORLD)
