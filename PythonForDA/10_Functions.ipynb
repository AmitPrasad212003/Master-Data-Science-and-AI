## 1ï¸âƒ£ What is a Function in Python?

A **function** is a **block of reusable code** that:

- Performs a specific task
- Runs **only when called**
- Helps reduce repetition
- Improves readability and maintainability

### Real-life example

Instead of writing the same logic again and again, you write it **once** and reuse it.

## 8ï¸âƒ£ Lambda Function (In-Depth)

### What is Lambda?

- Anonymous (no name)
- Single expression
- Automatically returns value
- Cannot contain statements

### Syntax

```python
lambda arguments: expression

```

---

### Lambda vs Normal Function

| Feature | def | lambda |
| --- | --- | --- |
| Name | Yes | No |
| Lines | Multiple | Single |
| Return | `return` | Implicit |
| Readability | High | Medium |

---

### Examples

### Single Argument

```python
square =lambda x: x * x

```

### Multiple Arguments

```python
add =lambda a, b: a + b

```

---

### Lambda with `map()`

```python
nums = [1,2,3]
print(list(map(lambda x: x *2, nums)))

```

---

### Lambda with `filter()`

```python
print(list(filter(lambda x: x %2 ==0, nums)))

```

---

### Lambda with `sorted()`

```python
pairs = [(1,3), (2,1)]
print(sorted(pairs, key=lambda x: x[1]))

```
