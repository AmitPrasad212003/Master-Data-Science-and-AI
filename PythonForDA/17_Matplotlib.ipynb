## 2ï¸âƒ£ `plt.plot()` â€“ Core Function

### Basic Syntax

```python
plt.plot(x, y)
plt.show()

```

---

### Example

```python
x = [1,2,3,4]
y = [10,20,30,40]

plt.plot(x, y)
plt.show()

```

### Default Output (IMPORTANT)

- Blue solid line
- No markers
- No title
- No axis labels
- Line width = 1
- Color = blue (`#1f77b4`)

ðŸ“ˆ Visually: a **simple blue line graph**

---

## 3ï¸âƒ£ Default Properties of `plt.plot()`

| Property | Default |
| --- | --- |
| Color | Blue |
| Line style | Solid (`-`) |
| Marker | None |
| Line width | 1 |
| Label | None |
| Grid | Off |

## 4ï¸âƒ£ Adding Labels & Title (Default Behavior)

### Output

## 5ï¸âƒ£ Combining MULTIPLE LINES in ONE `plot()` call (IMPORTANT ðŸ”¥)

### Syntax

```python
plt.plot(x1, y1, x2, y2)

```

---

### Example

```python
x = [1,2,3,4]
y1 = [10,20,30,40]
y2 = [5,15,25,35]

plt.plot(x, y1, x, y2)
plt.show()

```

### Default Output

- Line 1 â†’ Blue
- Line 2 â†’ Orange
- Both solid lines
- No legend (by default)

ðŸ“Œ Matplotlib **automatically assigns colors**

---

## 6ï¸âƒ£ Combining MULTIPLE LINES with Styles in ONE LINE

### Syntax

```python
plt.plot(x, y1,'r--', x, y2,'go-')

```

---

### Example

```python
plt.plot(x, y1,'r--', x, y2,'go-')
plt.show()

```

### Meaning

| Code | Meaning |
| --- | --- |
| `r--` | Red dashed line |
| `g` | Green |
| `o` | Circle marker |
| `-` | Solid line |

---

## 7ï¸âƒ£ Format String Syntax (VERY IMPORTANT)

```python
'color marker linestyle'

```

### Examples

| Format | Meaning |
| --- | --- |
| `'r-'` | Red solid |
| `'b--'` | Blue dashed |
| `'go'` | Green circles |
| `'k^:'` | Black triangle dotted |

---

## 8ï¸âƒ£ Multiple Lines with Labels (ONE LINE + LEGEND)

### Syntax

```python
plt.plot(x, y1, label="Line1")
plt.plot(x, y2, label="Line2")
plt.legend()

```

---

### Why legend not in one line?

âš ï¸ Labels must be assigned **per line**

| Method | Purpose |

| `plot()` | Line plot |

| `hist()` | Histogram |
| `pie()` | Pie |
| `xlabel()` | X label |
| `ylabel()` | Y label |
| `title()` | Title |
| `legend()` | Legend |
| `grid()` | Grid |
| `savefig()` | Save plot |
| `subplots()` | Multiple plots |
