## 1ï¸âƒ£ What are `loc` and `iloc`?

| Attribute | Meaning |
| --- | --- |
| `loc` | **Label-based indexing** |
| `iloc` | **Integer position-based indexing** |

---

## 2ï¸âƒ£ Sample DataFrame (Weâ€™ll use this everywhere)

```python
import pandasas pd
```

# ðŸ”¹ 3ï¸âƒ£ `loc` (Label-based)

### Rule

```python
df.loc[row_label, column_label]

```

âœ” Uses **index labels**

âœ” Uses **column names**

âœ” End index is **INCLUSIVE**

---

## Example 1: Select Single Row

```python
print(df.loc[101])

```

### Output

```
Name      A
Marks90
Age20
Name:101, dtype:object

```

---

## Example 2: Select Multiple Rows

```python
print(df.loc[101:103])

```

### Output

```
     NameMarks  Age
101A9020
102B8021
103     C8522

```

ðŸ“Œ **103 included** (VERY IMPORTANT)

---

## Example 3: Select Specific Columns

```python
print(df.loc[:, ["Name","Marks"]])

```

### Output

```
     NameMarks
101A90
102B80
103     C85
104     D70

```

---

## Example 4: Conditional Selection (POWERFUL)

```python

```

### Output

```
     NameMarks
101A90
103     C85

```

ðŸ“Œ `iloc` âŒ cannot do this directly

---

# ðŸ”¹ 4ï¸âƒ£ `iloc` (Position-based)

### Rule

```python
df.iloc[row_index, column_index]

```

âœ” Uses **integer positions (0-based)**

âœ” End index is **EXCLUSIVE**

---

## Example 1: Single Row

```python

```

### Output

```

Marks90
Age20
Name:101, dtype:object

```

---

## Example 2: Multiple Rows

```python

```

### Output

```

```

ðŸ“Œ `3` excluded

---

## Example 3: Rows & Columns

```python

```

### Output

```

```

# ðŸ”¥ MOST IMPORTANT DIFFERENCES

| Feature | `loc` | `iloc` |
| --- | --- | --- |
| Indexing type | Label-based | Position-based |
| Index type | Names / labels | Integers only |
| End index | Inclusive | Exclusive |
| Boolean filtering | âœ… Yes | âŒ No |
| Column names | Allowed | âŒ Not allowed |
| Python slicing rules | âŒ | âœ… |

## loc vs iloc

The main distinction between the two methods is:

- loc gets rows (and/or columns) with particular labels.

- iloc gets rows (and/or columns) at integer locations.

````python
s = pd.Series(list("abcdef"), index=[49, 48, 47, 0, 1, 2])
s
````

````python
s.loc[0]    # value at index label 0
````

````python
s.iloc[0]   # value at index location 0
````

````python
s.loc[0:1]  # rows at index labels between 0 and 1 (inclusive)
````

````python
s.iloc[0:1] # rows at index location between 0 and 1 (exclusive)
````

### What if you sort the Series

````python
s = s.sort_index()
````

````python
s
````

````python
s.iloc[0]
````

````python
s.loc[0]    # value at index label 0
````

````python
s.iloc[0]   # value at index location 0
````

````python
s.loc[0:1]  # rows at index labels between 0 and 1 (inclusive)
````

````python
s.iloc[0:1] # rows at index location between 0 and 1 (exclusive)
````
