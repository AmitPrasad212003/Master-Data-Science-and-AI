# VARIABLES

---

**Definition**

In Python, a variable is a **name (identifier)** that refers to an **object in memory**, not a fixed storage location.

Multiple variables can point to the **same object**.

**Example**

```python
a =10
b = a

```

Here, `a` and `b` both refer to the same object `10`.

---

## 2. Dynamic Typing

**Explanation**

- Variable type is decided **at runtime**
- Same variable can hold different data types at different times
- Type checking happens during execution

**Example**

```python
x =10
x ="Python"
x =3.14

```

---

## 3. Variable Binding and Rebinding

**Binding**

Occurs when a variable name is first associated with an object.

**Rebinding**

Occurs when the same variable name points to a new object.

**Example**

```python
x =5# binding
x =10# rebinding

```

The object `5` is not changed; `x` now points to `10`.

---

## 4. Scope of Variables (LEGB Rule)

**Explanation**

Python searches variables in this order:

1. Local
2. Enclosing
3. Global
4. Built-in

**Example**

```python
x =10# Global

defouter():
    x =20# Enclosing
definner():
        x =30# Local
print(x)
    inner()

outer()

```

---

## 5. Global Variables

**Explanation**

- Defined outside functions
- Can be read inside functions
- To modify them inside a function, `global` keyword is required

**Example**

```python
x =10

defupdate():
global x
    x =20

update()

```

---

## 6. Nonlocal Variables

**Explanation**

- Used in nested functions
- Refers to variables in enclosing scope
- Does not affect global scope

**Example**

```python
defouter():
    x =10
definner():
nonlocal x
        x =20
    inner()
print(x)

outer()

```

---

## 7. Mutability Concept

**Explanation**

- **Immutable objects** cannot be changed after creation
- **Mutable objects** can be changed without rebinding

**Example (Immutable)**

```python
x =10
x = x +1

```

**Example (Mutable)**

```python
lst = [1,2,3]
lst.append(4)

```

---

## 8. Object Identity

**Explanation**

- Each object has a unique identity in memory
- Multiple variables may share the same identity
- Identity remains constant during object lifetime

**Example**

```python
a =10
b =10
print(id(a) ==id(b))

```

---

## 9. Variable Lifetime

**Explanation**

- Variables exist while their objects are reachable
- Objects without references are removed by garbage collection
- Lifetime depends on scope and references

**Example**

```python
deffun():
    x =10
return x

```

Here, `x` exists only during function execution.

# KEYWORDS

---

## 1. Meaning of Keywords

**Definition**

Keywords are **reserved words** with predefined meaning in Python.

They **cannot be used as identifiers**.

**Example (Invalid)**

```python
if =10# Error

```

---

## 2. Control Flow Keywords

**Purpose**

Control decision-making and looping.

**Example**

```python
if x >0:
pass

```

---

## 3. Function and Class Keywords

**Purpose**

Used to define functions, classes, and return values.

**Example**

```python
defadd(a, b):
return a + b

```

---

## 4. Scope and Namespace Keywords

**Purpose**

Control variable visibility and scope resolution.

**Example**

```python
global x
nonlocal y

```

---

## 5. Exception Handling Keywords

**Purpose**

Handle runtime errors and maintain program flow.

**Example**

```python
try:
    x = int("abc")
except ValueError:
print("Error occurred")
finally:
print("Done")

```

---

## 6. Logical and Boolean Keywords

**Purpose**

Used for logical operations and truth evaluation.

**Example**

```python
if True and False:
print("Yes")

```

---

## 7. Module and Import Keywords

**Purpose**

Import modules and manage namespaces.

**Example**

```python
import math
from math import sqrtas s

```

---

## 8. Resource Management Keywords

**Purpose**

Ensure safe acquisition and release of resources.

**Example**

```python
with open("file.txt")as f:
    data = f.read()

```

---

## 9. Keyword Characteristics

**Key Points**

- Keywords are case-sensitive
- Part of Python grammar
- Cannot be redefined
- May change across Python versions
